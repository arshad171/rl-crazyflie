```python
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import plotly.express as px
```

```python
print(os.getcwd())
```

```python
SAVE_PATH = f"../results_perf_it_train_plt.png"
```

```python
models = {
    "b_dist-lstm-v-61": {"path" : f"../results-nav-b_dist-lstm-v-61/", "color": "purple"},
    # "dist-err-rew-wi-v-50": {"path" : f"../results-nav-dist-err-rew-w_i-v-50/plt/i{INIT}/", "color": "red"},
    "b_dist-err-rew-wi-v-51": {"path" : f"../results-nav-b_dist-err-rew-w_i-v-51/", "color": "orange"},
    # "dist-v-3": {"path" : f"../results-nav-dist-v-3/plt/i{INIT}/", "color": "blue"},
    "b_dist-v-31": {"path" : f"../results-nav-b_dist-v-31/", "color": "blue"},
    "v-21": {"path" : f"../results-nav-v-21/", "color": "green"},
}
```

```python
fig_r = px.line()
fig_r.update_layout(
    # title_text="attr",
    # title_xanchor="center",
    # title_x=0.5,
    xaxis_title="Timesteps",
    yaxis_title="Returns",
)

for mdl in models:
    try:
        event_acc = EventAccumulator(os.path.join(models[mdl]["path"], "logs"), size_guidance={"scalars": 0})
        event_acc.Reload()

        data = event_acc.Scalars("train/returns")

        df = pd.DataFrame(data)

        fig_r.add_scatter(
            x=df.index * (df.iloc[1]["step"] - df.iloc[0]["step"]),
            y=df.value,
            line=dict(color=models[mdl]["color"]),
            name=mdl,
        )
    except KeyError as ke:
        print(f"skipping. {mdl}")
        continue

fig_r.show()

fig_r.write_image(SAVE_PATH)
```
