```python
import plotly.figure_factory as ff
import plotly.graph_objects as go


import numpy as np
```

```python
x, y = np.meshgrid(np.arange(-1, 1, 0.22), np.arange(-1, 1, 0.22))
z = 1
vx = x / np.sqrt(x**2 + y**2 + (1 - z) ** 2)
vy = y / np.sqrt(x**2 + y**2 + (1 - z) ** 2)

ux = -vx
uy = -vy

fig = ff.create_quiver(
    x,
    y,
    ux,
    uy,
    name="opt. control",
)

fig.add_trace(
    go.Scatter(
        x=[0],
        y=[0],
        mode="markers",
        marker_size=12,
        name="dest (x=0, y=0)",
    )
)

fig.update_layout(
    title_text="Optimal Actions (XY plane)",
    xaxis_title="X",
    yaxis_title="Y",
    height=500,
    width=500,
)
fig.show()
```

```python
x, z = np.meshgrid(np.arange(-1, 1, 0.22), np.arange(0, 2, 0.22))
y = 0

vx = x / np.sqrt(x**2 + y**2 + (1 - z) ** 2)
vz = -(1 - z) / np.sqrt(x**2 + y**2 + (1 - z) ** 2)

ux = -vx
uz = -vz

fig = ff.create_quiver(
    x,
    z,
    ux,
    uz,
    name="opt. control",
)

fig.add_trace(
    go.Scatter(
        x=[0],
        y=[1],
        mode="markers",
        marker_size=12,
        name="dest (x=0, z=1)",
    )
)

fig.update_layout(
    title_text="Optimal Actions (XZ plane)",
    xaxis_title="X",
    yaxis_title="Z",
    height=500,
    width=500,
)
fig.show()
```
